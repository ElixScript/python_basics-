Dalam Python, `while` loop adalah struktur pengulangan yang memungkinkan Anda untuk menjalankan sekelompok pernyataan berulang kali selama suatu kondisi tertentu terpenuhi. Berikut adalah penjelasan detail mengenai `while` loop di Python:

### Sintaks
Sintaks dasar dari `while` loop adalah sebagai berikut:

```python
while kondisi:
    # blok kode yang akan diulang
```

- **`kondisi`**: Ini adalah ekspresi boolean yang dievaluasi sebelum setiap iterasi. Jika kondisi tersebut `True`, maka blok kode di dalam loop akan dijalankan. Jika kondisi tersebut `False`, eksekusi loop akan berhenti.
- **`blok kode yang akan diulang`**: Ini adalah kumpulan pernyataan yang akan dijalankan berulang kali selama kondisi terpenuhi.

### Contoh Sederhana

Berikut adalah contoh sederhana dari `while` loop yang mencetak angka dari 1 hingga 5:

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

Output:
```
1
2
3
4
5
```

Dalam contoh di atas:
- `i` mulai dengan nilai 1.
- Selama `i` kurang dari atau sama dengan 5, loop akan berjalan.
- `print(i)` mencetak nilai `i` saat ini.
- `i += 1` meningkatkan nilai `i` sebesar 1 pada setiap iterasi.

### Loop Tak Terbatas
Jika kondisi `while` selalu `True`, maka loop akan berjalan tanpa henti. Loop semacam ini disebut sebagai loop tak terbatas (infinite loop). Ini dapat dihentikan menggunakan `break` atau dengan menghentikan program secara manual.

Contoh infinite loop:
```python
while True:
    print("Ini adalah loop tak terbatas")
    break  # Menggunakan break untuk keluar dari loop
```

### Penggunaan `else` dengan `while`
Python juga mendukung penggunaan klausa `else` dengan `while` loop. Blok `else` akan dieksekusi jika loop berakhir secara normal (bukan karena `break`).

Contoh:
```python
i = 1
while i <= 5:
    print(i)
    i += 1
else:
    print("Loop selesai")
```

Output:
```
1
2
3
4
5
Loop selesai
```

### Penggunaan `break` dan `continue`
- **`break`**: Digunakan untuk menghentikan loop sebelum kondisi menjadi `False`.
- **`continue`**: Digunakan untuk melewati sisa dari blok kode di dalam loop dan melanjutkan ke iterasi berikutnya.

Contoh menggunakan `break`:
```python
i = 1
while i <= 5:
    if i == 3:
        break
    print(i)
    i += 1
```

Output:
```
1
2
```

Contoh menggunakan `continue`:
```python
i = 1
while i <= 5:
    i += 1
    if i == 3:
        continue
    print(i)
```

Output:
```
2
4
5
6
```

Dalam contoh di atas, angka 3 dilewati karena ketika `i == 3`, pernyataan `continue` menyebabkan loop melompat ke iterasi berikutnya tanpa menjalankan `print(i)` untuk `i == 3`.

Itulah dasar-dasar `while` loop di Python. Anda dapat menggunakannya untuk melakukan berbagai jenis pengulangan berdasarkan kondisi yang Anda tentukan.

````python
while False :
    print('hello')

# disini defaultnya adalah true
````
