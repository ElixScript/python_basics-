# law of large numbers

Hukum Bilangan Besar (Law of Large Numbers, LLN) adalah salah satu konsep fundamental dalam probabilitas dan statistik. Hukum ini menyatakan bahwa seiring dengan semakin banyaknya percobaan yang dilakukan, rata-rata hasil yang diperoleh akan semakin mendekati nilai ekspektasi teoretis. 
obabilitas 1.

### Contoh Ilustrasi
Misalkan kita melempar sebuah koin yang adil (probabilitas munculnya kepala adalah 0,5) sebanyak \(n\) kali dan mencatat hasilnya. Misalkan \(X_i\) adalah 1 jika hasil lemparan ke-\(i\) adalah kepala, dan 0 jika bukan kepala. Nilai ekspektasi dari setiap \(X_i\) adalah 0,5.

- Menurut Hukum Bilangan Besar Lemah, jika kita mengulang eksperimen tersebut berkali-kali, rata-rata jumlah kepala yang muncul akan mendekati 0,5.
- Menurut Hukum Bilangan Besar Kuat, rata-rata jumlah kepala yang muncul akan hampir pasti mendekati 0,5 seiring bertambahnalisis statistik yang andal.

```python
import numpy as np
from numpy.random import randn

N = 10000  # ini untuk ukuran sampelnya
counter = 0  # reset hitungannya

for x in randn(N):
    if (x > -1 and x < 1):  # dalam satu deviasi standar
        counter += 1

answer = counter / N
print(answer)
```
